```python
# Set percentage of column to 0 (pandas)
N = .2
total = len(dataset.index)

#convert to int for always integer
i = int(total * N)
idx = np.r_[0:i, total-i:total]

# Here i have added 1 in 20% records of a column, not considered where value is 1
dataset.iloc[idx, dataset.columns.get_loc('temp_condition')] = 1
```

```python
data_n = np.random.uniform(20, 26, size = 245000)
data_b = np.random.uniform(23, 28, size = 35000)
data_i = np.random.uniform(27, 32, size = 35000)
data_p = np.random.uniform(31, 36, size = 35000)
data_d = np.random.uniform(35, 40, size = 25000)
data_r = np.random.uniform(39, 43, size = 16030)
data_x = np.random.uniform(42, 47, size = 6116)
data_s = np.random.uniform(46, 48, size = 3973)
```

```python
print(data_s[0])
```

```python
dataset1.loc[dataset1['type'] == 'normal', 'temperature'] = data_n
dataset1.loc[dataset1['type'] == 'backdoor', 'temperature'] = data_b
dataset1.loc[dataset1['type'] == 'injection', 'temperature'] = data_i
dataset1.loc[dataset1['type'] == 'password', 'temperature'] = data_p
dataset1.loc[dataset1['type'] == 'ddos', 'temperature'] = data_d
dataset1.loc[dataset1['type'] == 'ransomware', 'temperature'] = data_r
dataset1.loc[dataset1['type'] == 'xss', 'temperature'] = data_x
dataset1.loc[dataset1['type'] == 'scanning', 'temperature'] = data_s
```

```python
print(dataset1.loc[dataset1['type'] == 'normal','temperature'].min())
print(dataset1.loc[dataset1['type'] == 'normal','temperature'].max())
```

```python
dataset2 = dataset1.copy()
```

```python
#rand_idx = dataset1[dataset1['type'] == 'normal'].index
```

```python
percentage = 15
arr = dataset2['temperature'].values

shuf = np.random.choice(np.arange(arr.shape[0]), round(arr.shape[0]*percentage/100), replace=False) 
arr[np.sort(shuf)] = arr[shuf] 

dataset2['temperature'] = arr
```

```python
print(dataset2.loc[dataset2['type'] == 'normal','temperature'].min())
print(dataset2.loc[dataset2['type'] == 'normal','temperature'].max())
```
