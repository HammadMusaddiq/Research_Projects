```python
import csv
import tqdm
from datetime import datetime
from numba import jit
```

```python
print('reading fridge data')

iot_fridge = []  # datetime,temperature,label
with open('Usman_IoT/IoT_Fridge.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in tqdm.tqdm(csv_reader):
        if line_count != 0:
            row[1] = row[1].replace(' ', '')
            date_time_obj = datetime.strptime(row[0] + ' ' + row[1], '%d-%b-%y %H:%M:%S')
            iot_fridge.append([date_time_obj, float(row[2]), int(row[4])])
        line_count = line_count + 1

print('fridge rows ' + str(len(iot_fridge)))
```

```python
print('reading garage door data')
iot_garage_door = []  # datetime,signal,label
with open('Usman_IoT/IoT_Garage_Door.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in tqdm.tqdm(csv_reader):
        if line_count != 0:
            try:
                row[1] = row[1].replace(' ', '')
                date_time_obj = datetime.strptime(row[0] + ' ' + row[1], '%d-%b-%y %H:%M:%S')
                iot_garage_door.append([date_time_obj, int(bool(row[3])), int(row[4])])
            except:
                pass
        line_count = line_count + 1
print('garage door rows ' + str(len(iot_garage_door)))
```

```python
print('reading gps tracker data')
iot_gps_tracker = []  # datetime,latitude,longitude,label
with open('Usman_IoT/IoT_GPS_Tracker.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in tqdm.tqdm(csv_reader):
        if line_count != 0:
            try:
                row[1] = row[1].replace(' ', '')
                date_time_obj = datetime.strptime(row[0] + ' ' + row[1], '%d-%b-%y %H:%M:%S')
                iot_gps_tracker.append([date_time_obj, float(row[2]), float(row[3]), int(row[4])])
            except:
                pass
        line_count = line_count + 1
print('gps tracker rows ' + str(len(iot_gps_tracker)))
```

```python
print('reading modbus data')
iot_modbus = []  # datetime,FC1_Read_Input_Register,FC2_Read_Discrete_Value,FC3_Read_Holding_Register,FC4_Read_Coil,
# label
with open('Usman_IoT/IoT_Modbus.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in tqdm.tqdm(csv_reader):
        if line_count != 0:
            try:
                row[1] = row[1].replace(' ', '')
                date_time_obj = datetime.strptime(row[0] + ' ' + row[1], '%d-%b-%y %H:%M:%S')
                iot_modbus.append(
                    [date_time_obj, float(row[2]), float(row[3]), float(row[4]), float(row[5]), int(row[6])])
            except:
                pass
        line_count = line_count + 1
print('modbus rows ' + str(len(iot_modbus)))
```

```python
print('reading motion light data')
iot_motion_light = []  # datetime,latitude,longitude,label
with open('Usman_IoT/IoT_Motion_Light.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in tqdm.tqdm(csv_reader):
        if line_count != 0:
            try:
                row[1] = row[1].replace(' ', '')
                date_time_obj = datetime.strptime(row[0] + ' ' + row[1], '%d-%b-%y %H:%M:%S')
                light = 0
                if row[3] == 'on':
                    light = 1
                iot_motion_light.append([date_time_obj, int(row[2]), light, int(row[4])])
            except:
                pass
        line_count = line_count + 1
print('motion light rows ' + str(len(iot_motion_light)))
```

```python
print('reading thermostate data')
iot_therostate = []  # datetime,temp,status,label
with open('Usman_IoT/IoT_Thermostat.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in tqdm.tqdm(csv_reader):
        if line_count != 0:
            try:
                row[1] = row[1].replace(' ', '')
                date_time_obj = datetime.strptime(row[0] + ' ' + row[1], '%d-%b-%y %H:%M:%S')

                iot_therostate.append([date_time_obj, float(row[2]), int(row[3]), int(row[4])])
            except:
                pass
        line_count = line_count + 1
print('thermostate rows ' + str(len(iot_therostate)))
```

```python
print('reading weather data')
iot_weather = []  # datetime,temp,status,label
with open('Usman_IoT/IoT_Weather.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in tqdm.tqdm(csv_reader):
        if line_count != 0:
            try:
                row[1] = row[1].replace(' ', '')
                date_time_obj = datetime.strptime(row[0] + ' ' + row[1], '%d-%b-%y %H:%M:%S')

                iot_weather.append([date_time_obj, float(row[2]), float(row[3]), float(row[4]), int(row[5])])
            except:
                pass
        line_count = line_count + 1
print('weather rows ' + str(len(iot_weather)))
```

```python
iot_fridge = sorted(iot_fridge, key=lambda k: k[0])
iot_garage_door = sorted(iot_garage_door, key=lambda k: k[0])
iot_gps_tracker = sorted(iot_gps_tracker, key=lambda k: k[0])
iot_modbus = sorted(iot_modbus, key=lambda k: k[0])
iot_motion_light = sorted(iot_motion_light, key=lambda k: k[0])
iot_therostate = sorted(iot_therostate, key=lambda k: k[0])
iot_weather = sorted(iot_weather, key=lambda k: k[0])
```

```python
selected_fridge = []
selected_garage = []
selected_gps = []
selected_modbus = []
selected_motionlight = []
selected_thermostate = []
selected_weather = []
```

```python
garage_index = 0
gps_index = 0
modbus_index = 0
motion_light_index = 0
thermostate_index = 0
weather_index = 0
```

```python
for fridge in tqdm.tqdm(iot_modbus):
    fridge_date = fridge[0]
    fridge_label = fridge[5]
    s_garage = None

    for i in range(garage_index, len(iot_garage_door)):

        if iot_garage_door[i][0] == fridge_date and fridge_label == iot_garage_door[i][2]:
            s_garage = iot_garage_door[i]
            garage_index = i

            break
        else:
            if fridge_date < iot_garage_door[i][0]:
                break

    s_gps = None
    for i in range(gps_index, len(iot_gps_tracker)):
        if iot_gps_tracker[i][0] == fridge_date and fridge_label == iot_gps_tracker[i][3]:
            s_gps = iot_gps_tracker[i]
            gps_index = i
            break
        else:
            if fridge_date < iot_gps_tracker[i][0]:
                break

    s_modbus = None
    for i in range(modbus_index, len(iot_fridge)):
        if iot_fridge[i][0] == fridge_date and fridge_label == iot_fridge[i][2]:
            s_modbus = iot_fridge[i]
            modbus_index = i
            break
        else:
            if fridge_date < iot_fridge[i][0]:
                break

    s_motion_light = None
    for i in range(motion_light_index, len(iot_motion_light)):
        if iot_motion_light[i][0] == fridge_date and fridge_label == iot_motion_light[i][3]:
            s_motion_light = iot_motion_light[i]
            motion_light_index = i

            break
        else:
            if fridge_date < iot_motion_light[i][0]:
                break

    s_thermostate = None
    for i in range(thermostate_index, len(iot_therostate)):
        if iot_therostate[i][0] == fridge_date and fridge_label == iot_therostate[i][3]:
            s_thermostate = iot_therostate[i]
            thermostate_index = i
            break
        else:
            if fridge_date < iot_therostate[i][0]:
                break

    s_weather = None
    for i in range(weather_index, len(iot_weather)):
        if iot_weather[i][0] == fridge_date and fridge_label == iot_weather[i][4]:
            s_weather = iot_weather[i]
            weather_index = i
            break
        else:
            if fridge_date < iot_weather[i][0]:
                break
    if s_garage is not None and s_gps is not None and s_modbus is not None and s_motion_light is not None and s_thermostate is not None and s_weather is not None:
        selected_fridge.append(s_modbus)
        selected_garage.append(s_garage)
        selected_gps.append(s_gps)
        selected_modbus.append(fridge)
        selected_motionlight.append(s_motion_light)
        selected_thermostate.append(s_thermostate)
        selected_weather.append(s_weather)
```
