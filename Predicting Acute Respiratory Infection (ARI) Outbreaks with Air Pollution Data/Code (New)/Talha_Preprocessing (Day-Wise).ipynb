```python
X = data_final1.drop(['Disease_Label'], axis =1)
X = X.values
X
```

```python
y = data_final1['Disease_Label']
y = y.values
y
```

```python
from imblearn.over_sampling import SMOTE 
smote = SMOTE()
from collections import Counter
```

```python
X_S, y_S = smote.fit_resample(X,y)
```

```python
print("Before Smote :", Counter(y))
print("After Smote :", Counter(y_S))
```

```python
print(X_S.shape)
print(y_S.shape)
```

```python
#Stratified K Fold split
kfold = StratifiedKFold(n_splits=10, random_state=42, shuffle=True)
print(kfold)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_sc, y, test_size=0.25, shuffle=True)
```

```python
print(X.shape)
print(y.shape)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
y_predict = rf.predict(x_test)
print(accuracy_score(y_test, y_predict))
print(classification_report(y_test, y_predict))
```

```python
cv_results = cross_val_score(rf, X_sc, y, cv=kfold, scoring='accuracy')
cv_results_mean = cv_results.mean()
cv_results_std = cv_results.std()
cv_results_name = 'RF'
print('%s: %f (%f)' % (cv_results_name, cv_results_mean, cv_results_std))
```

```python
#for confussion matrix
y_RF = cross_val_predict(rf, X, y, cv=kfold)
#Accuracy matrix
accu_RF = accuracy_score(y, y_RF)
print(accu_RF)
#Confussion matric
conf_RF = confusion_matrix(y, y_RF)
print(conf_RF)
#Classification Report
crep_RF = classification_report(y, y_RF)
print(crep_RF)
```
